# Use Regex

```python
year=[]
Brand=[]
Carname=[]
Mileage=[]
owner=[]
fuel=[]
Location=[]
for i in carname:
    regex=re.findall("(\d+)\s",str(i))
    year.append(regex[0])
for i in carname:
    regex=re.findall("(?:Tata|TATA|Maruti|MARUTI|HONDA|Honda|Hyundai|HYUNDAI)",str(i))
    Brand.append(regex[0])
for i in carname:
    regex=re.findall("(?:Tata|TATA|Maruti|MARUTI|HONDA|Honda|Hyundai|HYUNDAI)\s(.*)",str(i))
    Carname.append(regex[0])
    
for i in km:
    regex=re.findall("(\d+,?\d+)\skm",str(i))
    if regex:
         Mileage.append(regex[0])
    else:
         Mileage.append(np.nan)

for i in km:
    regex=re.findall("(\d\D+)\sOwner",str(i))
    if regex:
        owner.append(regex[0])
    else:
        owner.append(np.nan)
        
for i in km:
    regex=re.findall("Petrol|Diesel|CNGHR|CNGMH|CNGAP|CNGDL|CNGUP",str(i))
    if regex:
        fuel.append(regex[0])
    else:
        fuel.append(np.nan)

for i in location:
    regex=re.findall(r"(?:Hyderabad|New Delhi|Mumbai|Bengaluru|Chennai|Gurgaon|Noida|Pune|Thiruverkadu)",str(i))
    if regex:
        Location.append(regex[0])
    else:
        Location.append(np.nan)
```

```python
car={'year':year,'Brand':Brand,'Carname':Carname,'Mileage':Mileage,'owner':owner,'price':price,'fuel':fuel,'emi':emi,'Location':Location}
```

# Raw Data Frame

```python
a=pd.DataFrame(car)
a
```

```python
a["emi"]=a["emi"].apply(lambda x:x.replace(",","").replace("/month","").replace("₹",""))
a["price"]=a["price"].apply(lambda x:x.replace("₹","").replace(",",""))
a["Mileage"]=a["Mileage"].apply(lambda x:x.replace(",",""))
a["owner"]=a["owner"].apply(lambda x:x.replace("st","").replace('nd','').replace('rd',''))
```

```python
a["Location"]=a["Location"].fillna("Hyderabad")
```
